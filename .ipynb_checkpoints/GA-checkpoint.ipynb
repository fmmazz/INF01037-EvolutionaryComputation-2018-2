```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random, operator
import re
```

Nesta primeira célula, é feita a leitura e processamento da instância que será avaliada. 
A instância é transformada em uma matriz completa de números reais que representa a distância entre as cidades.

Para utilizar diferentes instâncias, basta colocar o nome da instância desejada no datasetName. Todas as instâncias presentes na pasta /instances/ já foram pré-processadas de modo a funcionar nos experimentos.

```python
datasetName = "instances/br17.atsp"

patternFileName = re.compile(r'instances/(.*).atsp')

matchFileName = re.search(patternFileName, datasetName)

if matchFileName:
    fileName = matchFileName.group(1)
    
print(fileName)

file = open(datasetName, "r")
dataset = []
cityListFile = []

numCities = file.readline()
numCities = int(numCities)

for line in file:
    
    cityListFile += list(filter(None, line.strip().split(" ")))
    #print("cityListFile: ")
    #print(cityListFile)
    #print(len(cityListFile))
        
while len(cityListFile) > 0:
    #print(len(cityListFile))
    dataset.append(cityListFile[0:numCities])
    del cityListFile[0:numCities]
#cityListFile += list(filter(None, line.strip().split(" ")))
       
#dataset.append(cityListFile)

dataset = [[float(y) for y in x] for x in dataset]
    
dataset = np.array(dataset)
print("Matriz de distância extraída do dataset %s:" % (datasetName))
print(dataset)

numCities = len(dataset)
print("\n----------\n")
print("Número de cidades do dataset %s: %d" % (datasetName, numCities))

cityList = [i for i in range(numCities)]
print("\n----------\n")
print("Lista de cidades do dataset " + datasetName + ": ",)
print(cityList)
```

------------------------------------------------------------------------------------------------------------

Na próxima célula é implementada a classe de Indivíduo. 

Cada instância da classe de indivíduo representa uma rota para o caixeiro viajante, ou seja, uma solução possível. 

O fitness ($1/distância$), a distância e a rota são atributos dessa classe. 

Além disso, essa classe possui dois métodos responsáveis por calcular a distância do caminho e o fitness da solução, respectivamente. 

------------------------------------------------------------------------------------------------------------

```python
class Individual:
    def __init__(self, path=[]):
        self.fitness= 0.0
        self.distance = 0
        self.path = path
        
    def createPath(self, cityList):
        cityList = cityList[1:]
        self.path = random.sample(cityList, len(cityList))
        self.path.insert(0, 0)
        return self.path
    
    
    def pathDistance(self):
        #print("!!!!!!!!")
        #print(self.path)
        #print("!!!!!!!!")
        
        if self.distance == 0:
            pathDistance = 0
            for i in range(0, len(self.path)):
                #print("i: %d" % i)
                originCity = self.path[i]
                destinationCity = None
                if i + 1 < len(self.path):
                    destinationCity = self.path[i + 1]
                else:
                    destinationCity = self.path[0]
                    
                pathDistance += dataset[originCity][destinationCity]
                    
                #print("\tOrigin city: %d" % originCity)
                #print("\tDestination city: %d" % destinationCity)
                #print("\tDistance from city %d to city %d: %d" % (originCity, destinationCity, dataset[originCity][destinationCity]))
                #print("\tCurrent path distance: %d" % pathDistance)
        
            self.distance = pathDistance
            #print("\tCurrent path distance: %d" % pathDistance)
        return self.distance
    
    def pathFitness(self):
        if self.fitness == 0:
            self.fitness = 1 / float(self.pathDistance())
        return self.fitness
```

------------------------------------------------------------------------------------------------------------
A terceira parte representa a classe de População. 

Essa classe é responsável por gerar a população inicial, classificar os as rotas pela sua distância, selecionar os indivíduos que irão criar a próxima geração, realizar o crossover dos indivíduos selecionados e realizar a mutação sobre a nova geração. 

Foi utilizada a Seleção Proporcional ao Fitness para selecionar os indivíduos que são escolhidos como pais. Essa seleção, também conhecida como Seleção da Roleta, utiliza o fitness relativo a população de cada um dos indivíduos para atribuir a probabilidade desse indivíduo ser selecionado. Para melhorar a seleção de bons indivíduos, também foi utilizado a característica de elitismo, em que os melhores indivíduos são sempre inseridos na nova geração de soluções. 

Para realizar o cruzamento dos pais selecionados, foi utilizada a técnica de Cruzamento Ordenado. Essa técnica foi escolhida pelo fato do formato da solução do problema ser uma lista sequencial de nodos. Esse método seleciona uma substring da rota do primeiro pai e preenche o restante da rota com genes do segundo pai, sem repetir nodos.

Para a operação de mutação, foram utilizadas três diferentes técnicas: Swap, Inversion e Scramble. O método de Swap escolhe dois nodos aleatórios da solução e troca a ordem de ambos. O método Scramble escolhe uma substring de tamanho aleatório da solução e mudam a ordem da substring de maneira aleatória. A técnica Inversion seleciona uma substring de tamanho aleatório da solução e inverte a ordem da substring, de maneira que a substring fique espelhada.

------------------------------------------------------------------------------------------------------------

```python
class Population:
    def __init__(self, population=[]):
        self.population = population
        self.fitnessRanking = {}
        self.selectedPaths = []
        self.nextPopulationGeneration = []
        
    def createInitialPopulation(self, populationSize, cityList):
        #print("\nCreating the initial population...\n")
        self.population = []

        for i in range(0, populationSize):
            ind = Individual()
            ind.createPath(cityList)
            self.population.append(ind)
        return self.population
        
    def classifyPaths(self):
        
        self.fitnessRanking = {}
        for i in range(0,len(self.population)):
            self.fitnessRanking[i] = self.population[i].pathFitness()
       
        return sorted(self.fitnessRanking.items(), key = operator.itemgetter(1), reverse = True)
    
    def selectIndividuals(self, rankingPopulation, eliteSize):

        self.selectedPaths = []
        df = pd.DataFrame(np.array(rankingPopulation), columns=["Individual","Fitness"])
        df['Cumulative_Sum'] = df.Fitness.cumsum()
        df['Cumulative_Percentual'] = 100 * df.Cumulative_Sum / df.Fitness.sum()

        for i in range(0, eliteSize):
            self.selectedPaths.append(rankingPopulation[i][0])
            
        for i in range(0, len(rankingPopulation) - eliteSize):
            pick = 100*random.random()
            
            for i in range(0, len(rankingPopulation)):
                if pick <= df.iat[i,3]:
                    if rankingPopulation[i][0] not in self.selectedPaths:
                        self.selectedPaths.append(rankingPopulation[i][0])
                    break
                    
        return self.selectedPaths
    
    def selectParents(self, selectedPaths):

        selectedParents = []
        for i in range(0, len(selectedPaths)):
            index = selectedPaths[i]
            selectedParents.append(self.population[index])
            
        return selectedParents
    
    def crossoverIndividuals(self, parent1, parent2):
        childPath = []
        childParent1 = []
        childParent2 = []

        gene1 = int(random.random() * len(parent1.path))
        if gene1 == 0:
            gene1 = 1

        gene2 = int(random.random() * len(parent1.path))
        if gene2 == 0:
            gene2 = 1

        firstGene = min(gene1, gene2)
        lastGene = max(gene1, gene2)

        for i in range(firstGene, lastGene):
            childParent1.append(parent1.path[i])
            
        childParent1.insert(0,0)

        childParent2 = [item for item in parent2.path if item not in childParent1 and item != 0]

        childPath = childParent1 + childParent2
        
        child = Individual(childPath)
        
        return child
    
    def crossoverPopulation(self, selectedParents, eliteSize):
        children = []
        
        parents = random.sample(selectedParents, len(selectedParents))

        for i in range(0,eliteSize):
            children.append(selectedParents[i])

        for i in range(0, len(selectedParents)):
            child = self.crossoverIndividuals(parents[i], parents[len(selectedParents)-i-1])
            children.append(child)
        
        return children
    
    def mutateSwap(self, individual, mutationRate):

        for gene in range(1, len(individual.path)):
            if(random.random() < mutationRate):
                geneToSwap = int(random.random() * len(individual.path))
                
                if geneToSwap == 0:
                    geneToSwap = 1                
                
                gene1 = individual.path[gene]
                gene2 = individual.path[geneToSwap]

                individual.path[gene] = gene2
                individual.path[geneToSwap] = gene1
                
        return individual
    
    def mutateInversion(self, individual, mutationRate):
        
        if(random.random() < mutationRate):
            
            gene1 = int(random.random() * len(individual.path))
            if gene1 == 0:
                gene1 = 1

            gene2 = int(random.random() * len(individual.path))
            if gene2 == 0:
                gene2 = 1

            firstGene = min(gene1, gene2)
            lastGene = max(gene1, gene2)
            
            individual.path[firstGene:lastGene] = individual.path[lastGene-1:firstGene-1:-1]
                
        return individual
    
    def mutateScramble(self, individual, mutationRate):
        
        # Scramble mutation is also used with permutation encoded chromosome. In this mutation, 
        # one has to pick a subset of genes at random and then randomly rearrange the alleles
        # in those positions. Subset does not have to be contiguous. 
        
        if(random.random() < mutationRate):
            
            gene1 = int(random.random() * len(individual.path))
            if gene1 == 0:
                gene1 = 1

            gene2 = int(random.random() * len(individual.path))
            if gene2 == 0:
                gene2 = 1

            firstGene = min(gene1, gene2)
            lastGene = max(gene1, gene2)
            
            scramble = individual.path[firstGene:lastGene]
            random.shuffle(scramble)
            
            individual.path[firstGene:lastGene] = scramble
        
        return individual        
    
    def mutateNextPopulationGeneration(self, population, mutationRate, typeMutation):
        populationAfterMutation = []
        
        #print("População que vai ser mutada: " + str(len(population)))

        for ind in range(0, len(population)):
            #print(ind)
            if typeMutation == "mutateScramble":
                individualAfterMutation = self.mutateScramble(population[ind], mutationRate)
            elif typeMutation == "mutateInversion":
                individualAfterMutation = self.mutateInversion(population[ind], mutationRate)
            elif typeMutation == "mutateSwap":
                individualAfterMutation = self.mutateSwap(population[ind], mutationRate)
                
            populationAfterMutation.append(individualAfterMutation)
            
        return populationAfterMutation
    
    def generateNextGeneration(self, eliteSize, mutationRate, typeMutation):
        popRanked = self.classifyPaths()
        selectionResults = self.selectIndividuals(popRanked, eliteSize)
        selectedParents = self.selectParents(selectionResults)
        children = self.crossoverPopulation(selectedParents, eliteSize)
        self.nextPopulationGeneration = self.mutateNextPopulationGeneration(children, mutationRate, typeMutation)
        self.population = self.nextPopulationGeneration

        return self.nextPopulationGeneration
```

------------------------------------------------------------------------------------------------------------

A última célula executa os experimentos. 

Cada instância foi executada utilizando as três diferentes técnicas de mutação. 

Os experimentos foram executados utilizando uma população inicial de tamanho 200, elite de tamanho 100, taxa de mutação de 20%, número de gerações igual a 200. 

Para cada técnica de mutação foram realizadas 10 iterações. 

Como resultado, foi escolhido o melhor valor obtido entre as 10 iterações. 

Para avaliar uma instância, basta colocar o nome da instância desejada na primeira célula e executar todas as células. A instância será avaliada utilizando as 3 diferentes técnicas de mutação. No final da execução de cada técnica de mutação, será exibido um gráfico mostrando a evolução da performance de cada iteração ao longo das gerações. 

------------------------------------------------------------------------------------------------------------

```python
def geneticAlgorithmPlot(population, popSize, eliteSize, mutationRate, generations, iteration, typeMutation):
    population.createInitialPopulation(popSize, population.population)
    #for i in range(len(population.population)):
        #print(str(i) + " : ", end='')
        #print(population.population[i].path)
    performance = []
    performance.append(1 / population.classifyPaths()[0][1])
    
    #print(population.nextGeneration(eliteSize, mutationRate))
    
    for i in range(0, generations):
        print("Geração " + str(i)) 
        population.generateNextGeneration(eliteSize, mutationRate, typeMutation)
        #print(1 / population.classifyPaths()[0][1])
        performance.append(1 / population.classifyPaths()[0][1])
        #print(progress)
        
    print("\n\n ====== Best result: ", end='')
    print(performance[-1])
    
    if typeMutation == "mutateScramble":
        plt.plot(performance, label="Scramble: run " + str(iteration))
    elif typeMutation == "mutateInversion":
        plt.plot(performance, label="Inversion: run " + str(iteration))
    elif typeMutation == "mutateSwap":
        plt.plot(performance, label="Swap: run " + str(iteration))
    
    
    return performance[-1]
    

bestResults = {'br17': 39, 'ft53': 6905, 'ft70': 38673, 'ftv33': 1286, 'ftv35': 1473, 'ftv38': 1530, 'ftv44': 1613, 'ftv47': 1776, 'ftv55': 1608, 'ftv64': 1839, 'ftv70': 1950, 'ftv90': 1579, 'ftv100': 1788, 'ftv110': 1958, 'ftv120': 2166, 'ftv130': 2307, 'ftv140': 2420, 'ftv150': 2611, 'ftv160': 2683, 'ftv170': 2755, 'kro124p': 36230, 'p43': 5620, 'rbg323': 1326, 'rbg358': 1163, 'rbg403': 2465, 'rbg443': 2720, 'ry48p': 14422}
resultsSwap = []
for i in range(0,10):
    print("\n\n-----Swap Iteração " + str(i))
    pop = Population(cityList)
    resultsSwap.append(geneticAlgorithmPlot(population=pop, popSize=200, eliteSize=100, mutationRate=0.2, generations=200, iteration=i+1, typeMutation="mutateSwap"))
    
plt.ylabel('Distance')
plt.xlabel('Generation')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()

resultsInversion = []
for i in range(0,10):
    print("\n\n-----Inversion Iteração " + str(i))
    pop2 = Population(cityList)
    resultsInversion.append(geneticAlgorithmPlot(population=pop2, popSize=200, eliteSize=100, mutationRate=0.2, generations=200, iteration=i+1, typeMutation="mutateInversion"))

plt.ylabel('Distance')
plt.xlabel('Generation')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()

resultsScramble = []
for i in range(0,10):
    print("\n\n-----Scramble Iteração " + str(i))
    pop3 = Population(cityList)
    resultsScramble.append(geneticAlgorithmPlot(population=pop3, popSize=200, eliteSize=100, mutationRate=0.2, generations=200, iteration=i+1, typeMutation="mutateScramble"))

plt.ylabel('Distance')
plt.xlabel('Generation')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()

print("Results Swap:", end='')
print(resultsSwap)
print("Results Inversion:", end='')
print(resultsInversion)
print("Results Scramble:", end='')
print(resultsScramble)

print("Best result ever found: ", end='')
print(bestResults[fileName])

```
